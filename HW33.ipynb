```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# dataframe
db = pd.read_excel('auto.xls')
num = [i+1 for i in range(20)]
db.index = num
print(db, '\n', type(db))
```

```python
fig = plt.figure()  # (figsize=(10, 5))
ax = fig.add_subplot()
plt.xlabel('вартість, $')
plt.xticks(rotation=45)
plt.ylabel('споживання палива, л на 100 км')
plt.title('Споживання палива в залежності від вартості') 
plt.plot(db['вартість'], db['розх_палива'], color='blue')
# plt.show() 
```

```python
plt.xlabel('максимальна швидкість, км/год')
plt.ylabel('споживання палива, л на 100 км')
plt.title('Споживання палива в залежності від швидкості') 
plt.scatter(db['макс_швидк'], db['розх_палива'], color='red')
```

```python
plt.xlabel('максимальна швидкість, км/год')
plt.ylabel('вантажопідйомність, кг')
plt.title('Вантажопідйомність в залежності від швидкості') 
plt.scatter(db['макс_швидк'], db['вантаж'], color='green')
```

```python
plt.xlabel('максимальна швидкість, км/год')
plt.ylabel('частота')
plt.title('Частотний розподіл авто за швидкістю') 
plt.hist(db['макс_швидк'], bins=5, rwidth=0.9, color='blue')
```

```python
# boxplot для відображення швидкостей марок з різних країн 
# (Не можу відобразити для різних марок, тому що датасет містить по одній. Тому додаю ще стовпчик з країнами) 

country = pd.Series(['Jap', 'Ger', 'Cze', 'Fra', 'Jap', 'Ger', 'Jap', 'Swe', 'Cze', 'Fra', 
                     'Fra', 'Ger', 'Jap', 'Ger', 'Ger', 'Ger', 'Ita', 'Kor', 'Kor', 'Chi'], index=num)
db['країна'] = country
print(db)

sns.set_theme (style='darkgrid')
sns.boxplot(x = 'країна', y = 'макс_швидк', data = db)
```

```python
# парний графік для відображення залежностей величин
sns.pairplot(db)
```
